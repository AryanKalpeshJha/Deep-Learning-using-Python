```python
def sgmd(x):
  eq = 1/1+(np.exp(-x))
  return eq
```

```python
def acq(y_pred,y_true):
  acry = y_pred.argmax(axis=1)==y_true.argmax(axis=1)
  return acry.mean()
```

```python
def mean_sq_error(y_pred,y_true):
  eq = ((y_pred - y_true)**2).sum()/(2*len(y_pred))
  return eq
```

```python
N = input.size
iterations = 5000
learning_rate = 0.01
input_size = 4
hidden_size = 2
output_size = 3
```

```python
results = pd.DataFrame(columns=["mse","accuracy"])
results
```

```python
np.random.seed(20)
```

```python
W1 = np.random.normal(scale=0.5,size = (input_size,hidden_size))
W2 = np.random.normal(scale=0.5,size=(hidden_size,output_size))
```

```python
bias = np.random.rand(1)
```

```python
print(W1.shape)
print(W2.shape)
print(bias.shape)
print(input.shape)
```

```python
Z1 = np.dot(input,W1)+bias
A1 = sgmd(Z1)
Z2 = np.dot(A1,W2)
A2 = sgmd(Z2)
```

```python
print(Z1.shape)
print(A1.shape)
print(Z1.shape)
print(A2.shape)
```

```python
E1 = A2 - species
```

```python
print(E1.shape)
```

```python
for i in range(iterations):
  Z1 = np.dot(input,W1)+bias
  A1 = sgmd(Z1)
  Z2 = np.dot(A1,W2)
  A2 = sgmd(Z2)
  mse = mean_sq_error(A2,np.array(species))
  acc = acq(A2,np.array(species))
  df2 = {"mse":mse,"accuracy":acc}
  results = results.append(df2,ignore_index=True)
  E1 = A2 - species
  dw2 = E1*A2*(1-A2)
  E2 = np.dot(dw2,W2.T)
  dw1 = E2*A1*(1-A1)
  W2_update = np.dot(A1.T,dw2)/N
  W1_update = np.dot(input.T,dw1)/N
  W2 = W2 - learning_rate*W2_update
  W1 = W1 - learning_rate*W1_update
```
